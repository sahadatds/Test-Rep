# 1. Inventory Price Calculator

Real-world use: Used in e-commerce dashboards for price summaries.
Task:
Write a function calculate_inventory_summary (prices) that:
 receives a list of product prices
 returns total_cost and average_cost
Example:
Input: [120, 250, 399, 150]
Output: (919, 229.75)

```python
# Take user input for prices
prices_input = input("Enter product prices separated by commas: ")
# List comprehension
prices = [float(price.strip()) for price in prices_input.split(",")]
# define calculate_inventory_summary and pass prices parameter
def calculate_inventory_summary(prices):
    total_cost = 0
    # Loop through each item in prices list
    for price in prices:
        total_cost += price
    #Calculating average cost
    average_cost = total_cost / len(prices)
    return (total_cost, average_cost)
# Calling calculate_inventory_summary function with arguments
final = calculate_inventory_summary(prices)
print(final)
```

# 2. Unique Visitor Tracker

Real-world use: Website analytics tools calculating daily active users.
Task:
Write a function count_unique_visitors(visitor_list) that:
 takes a list of visitor IDs
 returns the number of unique visitors
Example:
Input: [101, 205, 101, 310, 205, 550]
Output: 4 (The unique IDs are 101, 205, 310, 550)

```python
# Take user input
visitor_input = input("Enter visitor IDs separated by commas: ")
visitor_list = [int(visitor.strip()) for visitor in visitor_input.split(",")]
# Defining empty list unique_visitors
unique_visitors = []
# Defining count_unique_visitors and passing visitor_list parameter
def count_unique_visitors(visitor_list):
    #Loop through each item in visitor_list
    for visitor in visitor_list:
        # Check if visitor not in unique_visitors list
        if visitor not in unique_visitors:
            unique_visitors.append(visitor)
    return len(unique_visitors)
# Calling the function with visitor_list argument
visitor_count = count_unique_visitors(visitor_list)
# Print out the results in tuple format
print(visitor_count, "(The unique IDs are", ', '.join(str(x) for x in unique_visitors), ")")
```

# 3. Product Category Counter

```python
"""
    Real-world use: Inventory/category-based analytics.
Task:
Write a function count_categories(products) that:
 takes a list of product names
 returns a dictionary counting each product type.

Example:
Input: ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]
Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

"""

# Take user input for products
products_input = input("Enter product names separated by commas: ")
# Product list comprehension
products = [product.strip() for product in products_input.split(",")]
# Defining count_categories function with products parameter
def count_categories(products):
    # Define empty dictonary unique
    unique = {}
    # Loop through each items in products list
    for product in products:
        # check if produc not in unique dictionary
        if product not in unique:
            unique[product] = 1
        else:
            unique[product] += 1
    return unique
# Calling count_categories function with products argument
result = count_categories(products)
print(result)
```

# 4. Temperature Analytics

Real-world use: Health tech temperature screening systems.
Task:
Write a function analyze_temperatures(temp_list) that:
 separates high (> 38°C) and normal temps
 returns two lists: (high_list, normal_list)

Example:
Input: [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]
Output:
High List (>38°C): [39.1, 40.5]
Normal List (<38°C): [36.5, 38.0, 37.2, 38.0, 37.9]
Full Output Tuple: ([39.1, 40.5], [36.5, 38.0, 37.2, 38.0, 37.9])

```python
# Take user input for temperature list
temp_input = input("Enter temperatures separated by commas: ")
temp_list = [float(temp.strip()) for temp in temp_input.split(",")]
# Defining analyze_temperatures function with tem_list parameter
def analyze_temperatures(temp_list):
    # Initialize 2 empty list high_list, normal_list
    high_list = []
    normal_list = []
    # Loop through each item in temp_list
    for temp in temp_list:
        # Check if temperature is greater or not
        if temp > 38:
            high_list.append(temp)
        else:
            normal_list.append(temp)
    
    return high_list, normal_list
# Calling the function with temp_list arguments
high, normal = analyze_temperatures(temp_list)
# Prining out high, normal, and tuple of (high, normal) temperature
print("High List (>38°C):", high)
print("Normal List (≤38°C):", normal)
print("Full Output Tuple:", (high, normal))
```

# 5. Menu Price Lookup

Real-world use: Restaurant ordering apps.
Task:
Write a function get_price(menu, item) that:
 receives a dictionary menu and an item name
 returns the price if exists
 otherwise returns "Item not found"

Example 1: Item Found
 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
 Input (Item): "Muffin"
 Output: 4.00
Example 2: Item Not Found
 Input (Menu): {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
 Input (Item): "Donut"
 Output: "Item not found"

```python
# Take menu as dictionary input
menu_input = input("Enter menu items in format item:price, separated by commas: ")
menu = {}
# Loop through each menu_input and assign it into menu dictionary
for pair in menu_input.split(","):
    # Extract name, price from menu dictionary
    name, price = pair.split(":")
    # Set name, price in menu dictionary
    menu[name.strip()] = float(price.strip())
# Take item input from user
item = input("Enter item name to look up price: ").strip()
# Defining get_price funtion with menu, item parameter
def get_price(menu, item):
    if item in menu:
        return f"{menu[item]:.2f}"
    else:
        return "Item not found"
result = get_price(menu, item)
# Calling the function with item and menu argument
print(result)
```

# 6. Student Grading System

Real-world use: School result automation.
Task:
Write a function calculate_grade(score) that:
 returns "A+", "A", "A-", "F" based on numeric marks
Write another function grade_students(student_dict) to:
 take a dictionary of student:score
 return student:grade dictionary
Example:
Input (Scores Dictionary): {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
Output (Grades Dictionary): {"Alice": "A+", "Bob": "A", "Charlie": "F", "David": "A-"}

```python
# Taking Score as User input
scores_input = input("Enter student scores (Use this format: name:score, separated by commas): ")
score = {}
for pair in scores_input.split(","):
    name, number = pair.split(":")
    score[name] = int(number)
# Defining function to calculate grade
def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    else:
        return "F"

# def grade_students(student_dict):
def grade_students(student_dict):
    grades = {}
    for student, score in student_dict.items():
        grades[student] = calculate_grade(score)
    return grades
# Calling grade_students with score argument
result = grade_students(score)
print(result)
```

# 7. Discount Calculator

Real-world use: Online store discount engine.
Task:
Write a function apply_discounts(products) where:
 products is a list of tuples: (name, price, discount)
 return a list of updated prices after discount
Example:
Input (List of Tuples): [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
Output (List of Prices): [64.00, 22.50, 285.00]

```python
products = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
def apply_discounts(products):
  discount_price = []
  for product in products:
    #Getting Price from Product
    price = product[1]
    #Calculating discount price
    dis_price = price * product[2] / 100
    #Now pushing discount price into discount_price list
    discount_price.append(price - dis_price)
  #Printing discount price
  return discount_price
    
# Caing the function
dis_price = apply_discounts(products)
print([f"{p:.2f}" for p in dis_price])
```

# 9. User Authentication Mock

Real-world use: Login systems (basic level).
Task:
Write a function login(username, password) that:
 checks if username exists
 checks if password matches
 returns "Login Successful" or "Invalid Credentials"

Example Setup (Mock Database): Assume the following credential database is used internally:
CREDENTIALS = {"alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}
Example 1: Successful Login
 Input (Username): "maria_t"
 Input (Password): "password101"
 Output: "Login Successful"
Example 2: Failed Login (Wrong Password)
 Input (Username): "alex_s"
 Input (Password): " securepwd12"
 Output: "Invalid Credentials"
Example 3: Failed Login (Non-existent User)
 Input (Username): "peter_z"
 Input (Password): "any_password"
 Output: "Invalid Credentials"

```python
# user credential database
CREDENTIALS = { "alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}
# Taking username and password as input
username = input("Enter Username: ")
password = input("Enter Password: ")
#Defining login Function
def login(username, password):
    # check if username in Credentials or not
    if username in CREDENTIALS:
        if password == CREDENTIALS[username]:
            return "Login Successful"
        else:
            return "Invalid Credentials."
    else:
        return "Invalid Credentials"
# Calling the function and store it's value to result variable
result = login(username, password)
# Printing result
print(result)
```

# 10. Shopping Cart Bill Calculator

Real-world use: Billing systems in e-commerce apps.
Task:
Write a function calculate_total(cart, prices) that:
 takes a shopping cart list
 takes a dictionary of item:price
 returns the total bill amount
Milk, Bread, Milk, Cheese, Bread
Milk: 3.00, Bread: 2.50, Eggs: 4.00, Cheese: 5.00
Example:
Input (Cart): ["Milk", "Bread", "Milk", "Cheese", "Bread"]
Input (Price): {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}
Output (Total Bill Amount): 16.00

```python
# shopping cart list
cart_input = input("Enter your cart item ',' Seperate each item with comma. ")
cart = [ item.strip() for item in cart_input.split(",")]
# dictionary of item:price
prices_input  = input("Enter your price item ',' Seperate each item with comma. ")
prices = {}
# Loop throught each price item
for pair in prices_input.split(","):
  item, price = pair.split(":")
  prices[item.strip()] = float(price.strip())
# defining the function calculate_total(cart, prices)
def calculate_total(cart, prices):
  total_bill = 0
  for item in cart:
      total_bill += prices[item]
  return total_bill
#Calling the function and store the result in bill variable
bill = calculate_total(cart, prices)
# Print out the result
print(f"{bill:.2f}")
```
