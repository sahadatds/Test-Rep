# 1 Email Validator (String Filtering + Loops)

üíº Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.
Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
email = input("Please write your email : ")
# if '@' in email and (email.endswith('.com') or email.endswith('.org')):
#   print('Valid email')
# else:
#   print('Invalid Email. Please provide valid email.')

#check for @ using for loop
has_at = False
for i in email:
  if i == '@':
    has_at = True
    break

#check for valid extension
has_ext = False
if len(email) >= 4:
  ext = email[-4:]
  if ext == '.com' or ext == '.org':
    has_ext = True

#Validate
if has_ext and has_at:
  print("Valid Email")
else:
  print('Invalid Email. Please provide valid email.')
```

# 2. Count Word Frequency in Customer Reviews

üíº Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
review = input("Enter review : ")

#convert to lowercase for case sensitive
review_lower = review.lower()

#split into words
words = review_lower.split()

#count word frequencies
word_counts = {}

for word in words:
  if word in word_counts:
    word_counts[word] += 1
  else:
    word_counts[word] = 1

#display the results
print(word_counts)
```

# 3. Password Strength Checker

üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python
password = input("Enter your password: ")

#Initialize Flags
has_uppercase = False
has_lowercase = False
has_digit = False
has_special = False

#define special charecters
special_chars = "!@#$%^&*()-_+=[]{}|;:',.<>?/`~"

# loop through each charecter in password
for char in password:
  #check if uppercase
  if char >= 'A' and char <= 'Z':
    has_uppercase = True
  elif char >= 'a' and char <= 'z':
    has_lowercase = True
  elif char >= '0' and char <= '9':
    has_digit = True
  elif char in special_chars:
    has_special = True

#Print Result
if has_uppercase and has_lowercase and has_special:
  print("Strong Password")
else:
  print("Weak Password")
```

# 4. Username Generator

üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamitha57

```python
import random
# Get full name
full_name = input("Enter your full name: ")

# Split name by space
first_name = ""
last_name = ""
found_space = False

for char in full_name:
    if char == ' ':
        found_space = True
    elif found_space:
        last_name += char
    else:
        first_name += char

# Convert to lowercase
first_lower = ""
for char in first_name:
    if char >= 'A' and char <= 'Z':
        first_lower += chr(ord(char) + 32)
    else:
        first_lower += char

last_lower = ""
for char in last_name:
    if char >= 'A' and char <= 'Z':
        last_lower += chr(ord(char) + 32)
    else:
        last_lower += char

# First 3 of first name
first_3 = ""
for i in range(3):
    first_3 += first_lower[i]

# Last 3 of last name
last_3 = ""
for i in range(len(last_lower) - 3, len(last_lower)):
    last_3 += last_lower[i]

# Random number
random_num = random.randint(0, 100)

# Create username
username = first_3 + last_3 + str(random_num)

print(username)
```

# 5. SMS Word Counter

üíº Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit

```python
# message input
message = input("Enter your message: ")

# Count characters
char_count = 0
for char in message:
    char_count += 1

# Count words
word_count = 0
for i in range(len(message)):
    if message[i] == ' ':
        word_count += 1
word_count += 1

# Check if exceeds 160 characters
if char_count > 160:
    status = "Exceeds Limit"
else:
    status = "Within Limit"

# Print result
print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```

# 6. Remove Duplicates from Contact List

üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
# Get input
contacts = input("Enter names: ")

# Split by comma
names = contacts.split(", ")

# Remove duplicates
unique = []
for name in names:
    if name not in unique:
        unique.append(name)

# Print
print(", ".join(unique))
```

# 7. Detect Suspicious Words in Text (Moderation)

üíº Real-world use: Used in chat moderation or email filtering.
Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
chat = input("Enter text: ")
banned = ['spam', 'scam', 'fake']

found = False
for word in banned:
    if word in chat.lower():
        found = True
        break

if found:
    print("Warning: Suspicious content detected")
else:
    print("Text is safe")
```

# 8. Invoice Code Formatter

üíº Real-world use: Formatting IDs or codes in finance or ERP systems.
Task:
Given an invoice number like 23, generate code in the format INV00023.

Output: INV00023

```python
# Get invoice number
invoice_num = input("Enter invoice number: ")

# Format: INV + zeros + number
formatted = "INV" + "0" * (5 - len(invoice_num)) + invoice_num

print(formatted)
```

# 10. Word Reversal for Data Encryption

üíº Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
# Get sentence
sentence = input("Enter sentence: ")

# Split into words
words = sentence.split()

# Reverse each word
reversed_words = []
for word in words:
    reversed_words.append(word[::-1])

# Join back
result = " ".join(reversed_words)

print(reversed_words)
```
