# 10. Word Reversal for Data Encryption

ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
text = input("Enter text: ")
words = text.split()

reverse_word = []
for word in words:
  reverse_word.append(word[::-1])

result = " ".join(reverse_word)
print(result)
```
