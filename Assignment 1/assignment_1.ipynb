# Assignment 1: Temperature Check

Problem:
Write a program that takes the current temperature as input and prints:
● "It's too cold!" if below 10°C
● "It's cool outside" if between 10°C and 25°C
● "It's hot outside!" if above 25°C

```python
# taking tempareture input from user
current_temp = int(input("Please write today's temerature in °C : "))
# logic building
if current_temp < 10:
  print("It's too cold!")
elif current_temp > 10 and current_temp < 25:
  print("It's cool outside")
elif current_temp > 25:
  print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System

Problem:
Ask the user for their account balance and the amount they want to withdraw.
● If the withdrawal amount is greater than the balance → print "Insufficient
Balance"
● If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.

```python
# Asking user about his account balance and how much he wants to withdraw
account_bal = int(input("How much balance is in your account? "))
withdraw_bal = int(input("How much taka you want to withdraw from your account? "))

#checking if the balance is sufficient or not
if withdraw_bal > account_bal:
  print("Insufficient Balance")
elif withdraw_bal <= account_bal:
  print("Transaction Successful")
  remaining_bal = account_bal - withdraw_bal
  print("Remaining Balance : ",remaining_bal)
```

# Assignment 3: Student Grade Calculator

Problem:
Input marks (0–100) and print grade based on:
● 90–100 → A+
● 80–89 → A

● 60–69 → C
● Below 60 → Fail

```python
# asking student to input his marks
marks = int(input("What's your mark? "))

#now checking his mark with grades
if marks >= 90 and marks <= 100:
  print("A+")
elif marks >= 80 and marks <= 89:
  print("A")
elif marks >= 70 and marks <= 79:
  print("B")
elif marks >= 60 and marks <= 69:
  print("C")
elif marks < 60:
  print("Fail")

```

# Assignment 5: Restaurant Bill with Discount

Problem:
Take total bill amount from the user:
● If the bill is above 1000 → apply 10% discount
● Otherwise → no discount
Then print the final bill amount.

```python
# taking total bill amount from user
total_bill = int(input("Tell me the toal bill amount : "))

# now make compare with bill
if total_bill > 1000 :
  discount = total_bill * .10
  print("Final Bill : ",total_bill - discount, "Taka")
else:
  print("No discount. Your Final Bill : ",total_bill, "Taka")
```

# Assignment 8: Leap Year Checker

Problem:
Take a year as input and check:
● If it is divisible by 400 → Leap Year
● Else if divisible by 4 but not by 100 → Leap Year
● Otherwise → Not a Leap Year

```python
# taking year as input
year = int(input("Year : "))

#now check if it's leap year or not
if year / 400:
  print("Leap Year")
elif year / 4 and not year / 100:
  print("Leap Year")
else:
  print("Not Leap Year.")
```

# Assignment 10: Electricity Bill Calculator

Problem:
Input number of units consumed and calculate bill:
● Up to 100 units → 5 BDT per unit
● 101–200 units → 7 BDT per unit
● Above 200 units → 10 BDT per unit
Print total payable amount.

```python
# taking num of units consumed as user input
units_consumed = int(input("Number of Unites Consumed : "))

#now calculating bill
if units_consumed <= 100:
  print("5 BDT per unit. Total Payable amount is : ", units_consumed * 5, "Taka")
elif units_consumed > 101 and units_consumed < 200:
  print("7 BDT per unit Total Payable amount is : ", units_consumed * 7, "Taka")
elif units_consumed > 200:
  print("10 BDT per unit Total Payable amount is : ", units_consumed * 10, "Taka")
```
